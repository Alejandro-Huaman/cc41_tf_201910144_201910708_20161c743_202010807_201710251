#Print Map

```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#READ .CSV
#Almacenes
almacenx, almaceny = [], []
with open('almacenes.csv', 'r', newline="") as almacenesFile:
  reader = csv.reader(almacenesFile, delimiter=";")
  almacenx.clear()
  almaceny.clear()
  
  for row in reader:
    almacenx.append(int(row[0]))
    almaceny.append(int(row[1]))

#puntos de entrega
pentregax, pentregay = [], []
with open('puntos_entrega.csv', 'r', newline="") as almacenesFile:
  reader = csv.reader(almacenesFile, delimiter=";")
  pentregax.clear()
  pentregay.clear()
  for row in reader:
    pentregax.append(int(row[0]))
    pentregay.append(int(row[1]))
```

```python
plt.figure(num=1, figsize=(40,40))
plt.scatter(almacenx,almaceny,color="blue", label="Puntos de distribuci√≥n")
plt.scatter(pentregax,pentregay,color="red", label="Puntos de entrega")

plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
