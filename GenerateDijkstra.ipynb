```python
import urllib as ul
import math
import heapq as hq 
```

```python
def readAdjl(fn):
  with ul.request.urlopen(fn) as f:
    L = []
    for line in f:
        L.append(list(map(int, line.strip().split())))
  return L
```

```python
def dijkstra(G, s):
  n = len(G)
  visited = [False]*n
  path = [None]*n
  cost = [math.inf]*n
  cost[s] = 0
  queue = [(s, 0)]
  while queue:
    u, g_u = hq.heappop(queue)
    if not visited[u]:
      visited[u] = True
      for v in G[u]:
        f = g_u + 1
        if f < cost[v]:
          cost[v] = f
          path[v] = u
          hq.heappush(queue, (v, f))

  return path, cost
```

```python
url = 'https://raw.githubusercontent.com/Alejandro-Huaman/cc41_tf_201910144_201910708_20161c743_202010807_201710251/main/Lista%20de%20Adyacencia/adjList.in'
G = readAdjl(url)
```

```python
path, cost = dijkstra(G, 0)

#Solo imprimirmos los 2000 primeros nodos con su respectivo padre
for i in range(2000):
  print(f"{i}: {path[i]}")
```
